```python
import pandas as pd
import numpy as nm
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sn
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("C:/Users/RUSHI/Day 21/general_data.csv")
data.head()
```

```python
data.dropna(inplace=True)
data.shape
```

```python
data = data.drop(columns = ['EmployeeID', 'EmployeeCount', 'Over18', 'StandardHours', ])
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

data.Attrition = enc.fit_transform(data.Attrition)

data['BusinessTravel'] = enc.fit_transform(data['BusinessTravel'])
data['Department'] = enc.fit_transform(data['Department'])
data['EducationField'] = enc.fit_transform(data['EducationField'])
data['Gender'] = enc.fit_transform(data['Gender'])
data['JobRole'] = enc.fit_transform(data['JobRole'])
data['MaritalStatus'] = enc.fit_transform(data['MaritalStatus'])
```

```python
data
```

```python
data.info()
```

```python
y = data['Attrition']
x = data.drop(columns = ['Attrition'])
```

```python
X = sn.add_constant(x)
logistic_model1 = sn.Logit(y,X)
```

```python
result1 = logistic_model1.fit()
```
